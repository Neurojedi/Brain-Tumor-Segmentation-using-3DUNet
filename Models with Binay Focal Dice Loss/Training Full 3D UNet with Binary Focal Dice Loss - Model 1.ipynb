## Training the Models

Now, I can finally start training the model. I will train three models using the `sm.losses.binary_focal_dice_loss()` with 0.001, 0.0001, and 0.00001 learning rates.

### Model 1 : lr = 0.001

```python
tf.keras.backend.clear_session()
model_bfocaldice = util.unet_model_3d(loss_function=sm.losses.binary_focal_dice_loss, initial_learning_rate=0.001,  deconvolution=False, 
                               batch_normalization=False, metrics=['accuracy',dice_coefficient, sm.metrics.IOUScore(threshold=0.5)])
```

```python
checkpoint = ModelCheckpoint(
    'fullmodel1_bfocaldice_lr0001.h5', 
    monitor='val_loss', 
    verbose=1, 
    save_best_only=True, 
    save_weights_only=False,
    mode='auto'

)

history=model_bfocaldice.fit_generator(generator=train_generator,
        epochs=40,
        validation_data=valid_generator,
        callbacks=[checkpoint])
```

```python
sns.set_theme()
pd.DataFrame(history.history).plot(figsize=(15, 10))
plt.xlabel("Epoch")
plt.legend(loc="best")
plt.show()
```
